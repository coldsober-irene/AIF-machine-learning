##artificial neural network

```python
import numpy as np
from random import choice
import tensorflow as tf
```

```python
def genFeature():
  return [choice([j for j in range(i)]) for i in range(1, 21)], int(sum([choice([j for j in range(i)]) for i in range(1, 20)]) / 80)
y = []
x = [(genFeature()[0], y.append(genFeature()[1])) for k in range(1)]
c = [x[0][0] for row in x]
features = tf.constant(c, dtype = tf.float32)
labels = tf.constant(y, dtype = tf.int8)
```

##shuffle

```python
tf.random.set_seed(42)
features = tf.random.shuffle(features)
features = tf.transpose(features)
labels = tf.random.shuffle(labels)
w = tf.random.Generator.from_seed(42)
weights = w.normal(shape = (15, 20))
```

```python
weights.shape, features.shape
```

```python
t = tf.matmul(weights, features)
t
```

##reLu

```python
reLu_results = [0 if j.numpy()[0] <= 0 else j.numpy()[0] for j in t]
reLu_results
```
