```python
from google.colab import drive
drive.mount("/content/drive")
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score # for measuring model performance
```

```python
# break the limitation of number of rows to indicate in a frame.
pd.set_option("max_rows", None)
# break the limitation of number of columns to indicate in a frame.
pd.set_option("max_columns", None)
```

```python
# change directory
import os
# os.listdir("/content/drive/MyDrive/AIF progress/machine learning/data")
os.chdir("/content/drive/MyDrive/AIF progress/machine learning/data")
```

```python
# read excel file with pandas
df = pd.read_excel(os.listdir()[0])
```

```python
df.columns = df.iloc[1]
df
#df = df.iloc[8:] # remove first 7 rows which are empty
#df
```

```python
# fill NAN with 0
df = df.fillna(0)
df
```

```python
# convert date into numerical data
encoder = LabelEncoder()
for row in range(len(df)):
  df.iloc[row]['Time'] = str(df.iloc[row]['Time'])
# create a new column of digit formed from time label
df["New Time"] = encoder.fit_transform(df["Time"])
```

```python
df
```

```python
# remove NAN value from the dataframe
df = df.fillna(0)[2:]
```

```python
# modify columns
cols = list(df.columns)
cols
cols[cols.index("Motor RPM")] = "Motor RPM 1"
cols[cols.index("Screw RPM")] = "Screw RPM 1"
df.columns = cols
```

```python
df
```

```python
# choose datasets
x_train = df.iloc[0:60,[-1, 5]][6:]
y_train = df.iloc[0:60,6][6:]
x_test = df.iloc[60:,[-1, 5]][6:]
y_test = df.iloc[60:,6][6:]
```

```python
# feed data to the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
pred_screwSpeed = model.predict([[5, 1500]])
```

```python
pred_screwSpeed
```

```python
# measure the performance of the model
performance = r2_score(y_test, model.predict(x_test))
print("performane: ", performance)
```
